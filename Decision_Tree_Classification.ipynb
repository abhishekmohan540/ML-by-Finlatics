Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Loading the dataset

```python
dataset = pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X
```

```python
y
```

Splitting data into Training and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=1)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
print(X_test)
```

Training the Decision tree clsssification model

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion ='entropy', random_state=0)
```

```python
classifier.fit(X_train,y_train)
```

Predicting for new data

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

Predicting the Test data

```python
y_pred = classifier.predict(X_test)
```

```python
print(type(y_pred))
```

```python
print(type(y_test))
```

```python
y_test = y_test.to_numpy()
```

```python
# Reshape the NumPy arrays
y_pred = y_pred.reshape(len(y_pred), 1)
y_test = y_test.reshape(len(y_test), 1)

# Now concatenate
result = np.concatenate((y_pred, y_test), axis=1)

# Print the result
print(result)
```

```python
## [0,0] and [1,1] represent correct prediction and rest combination represnrt incorrect prediction
```

Making the Confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
## 48 is TP , 35 is TN
```

```python
accuracy_score(y_test, y_pred)
```
